### Limpeza e tratamento de dados usando Sweetiz

```python
#importando pandas
import pandas as pd
```

```python
#importar dados
dataset = pd.read_csv("Churn.csv", sep=";")
#visulizar
dataset.head()
```

```python
#Dar nomes as colunas
dataset.columns = ["Id","Score","Estado","Genero","Idade","Patrimonio","Saldo","Produtos","TemCartCredito",
                    "Ativo","Salario","Saiu"]
```

```python
#visulizar
dataset.head()
```

```python
#importando o pacote do sweetviz
import sweetviz as sv
```

```python
#criando o report de visualizaÃ§Ã£o
analyze_report = sv.analyze(dataset)
analyze_report.show_html('analyze.html', open_browser=False)
```
